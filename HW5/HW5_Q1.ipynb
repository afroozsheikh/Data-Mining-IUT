```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Mining/HW5/Hackathon_Working_Data.csv')
df
```

# Part b

there is no missing value

```python
df.dropna(inplace=True)
df
```

# Part c

```python
df['Trans_ID'] = df['STORECODE'] + '_' + df['BILL_ID']
df.head()
```

# Part d

```python
df2 = (df.groupby(['Trans_ID', 'GRP'])['QTY'].sum().unstack().fillna(0).reset_index().set_index('Trans_ID'))
df2
```

# Part e

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

main_df = df2.applymap(encode_units)
main_df.dropna(inplace=True)
```

# Part f

```python
frequent_itemsets = apriori(main_df, min_support=0.01, use_colnames=True)
print(frequent_itemsets)
```

# Part g

```python
rules = association_rules(frequent_itemsets, metric="lift")
rules.head()
```
