# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_credit = pd.read_csv('/content/drive/MyDrive/Data Mining/HW2/credit cards.csv')
df_credit.head()
```

# Part a

```python
fig, ax = plt.subplots(figsize=(18, 10))
corr = df_credit.corr()
sns.heatmap(corr, annot = True, ax=ax)
```
